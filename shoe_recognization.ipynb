```python
import pandas as pd
df=pd.read_csv(r"/content/shoe/data/fashion.csv")
```

```python
df.head()
```

```python
shoedf=df[df["Category"].str.contains("Apparel")==False]
```

```python
shoedf.reset_index()
```

```python
shoedf.head()
```

```python
y = shoedf['Gender']
```

```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(shoedf, y, test_size=0.2)
```

```python
x_train, x_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.125)
```

```python
x_train.info()
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rotation_range = 30,
                                   zoom_range = 0.2,
                                   width_shift_range = 0.1,
                                   height_shift_range = 0.1,
                                   horizontal_flip = True,
                                   rescale = 1./255) #generate
val_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)
#batch_size = 100
#training_set = train_datagen.flow_from_directory(train_dir,target_size = (64, 64), batch_size = batch_size) #label
#val_set = val_datagen.flow_from_directory(validate, target_size = (64, 64), batch_size = batch_size)
#test_set = test_datagen.flow_from_directory(test_dir,
                                            #target_size = (64, 64),
                                            #batch_size = batch_size)
training_set=train_datagen.flow_from_dataframe(dataframe=x_train, directory="/content/shoe/data/Footwear", x_col="Image", y_col="Gender", class_mode="categorical", target_size=(64,64), batch_size=50)
val_set=val_datagen.flow_from_dataframe(dataframe=x_val, directory="/content/shoe/data/Footwear", x_col="Image", y_col="Gender", class_mode="categorical", target_size=(64,64), batch_size=50)
test_set=test_datagen.flow_from_dataframe(dataframe=X_test, directory="/content/shoe/data/Footwear", x_col="Image", y_col="Gender", class_mode="categorical", target_size=(64,64), batch_size=50)
```

```python
from keras.models import Sequential 
from keras.layers import Dense, Dropout, Flatten, BatchNormalization
from keras.layers import Conv2D, MaxPooling2D 

cnn = Sequential()
cnn.add(Conv2D(filters=32, kernel_size=3, activation='relu', input_shape=(64, 64, 3)))
cnn.add(BatchNormalization())
cnn.add(MaxPooling2D(pool_size=2, strides=2, padding='SAME'))
cnn.add(Conv2D(filters=64, kernel_size=3, activation='relu'))
cnn.add(BatchNormalization())
cnn.add(MaxPooling2D(pool_size=2, strides=2, padding='SAME'))
cnn.add(Flatten())
cnn.add(Dense(64, activation='relu'))
cnn.add(Dropout(0.25))
cnn.add(Dense(1, activation='sigmoid'))
```

```python
from keras import losses

cnn.compile(optimizer = 'adam', 
            loss = losses.BinaryCrossentropy(), 
            metrics = ['accuracy'])

history = cnn.fit(training_set, validation_data = val_set, epochs = 10) #
```

```python
test_loss, test_accuracy = cnn.evaluate(test_set)
```
