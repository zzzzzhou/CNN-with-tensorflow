```python
import numpy as np
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(shoedf, y, test_size=0.2)
```

```python
x_train, x_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.125)
```

```python
from keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(rotation_range = 30,
                                   zoom_range = 0.2,
                                   width_shift_range = 0.1,
                                   height_shift_range = 0.1,
                                   horizontal_flip = True,
                                   rescale = 1./255) #在范围内随机移动修改图像，在送入模型训练前进行generate
val_datagen = ImageDataGenerator(rescale = 1./255)
test_datagen = ImageDataGenerator(rescale = 1./255)
#batch_size = 100
#training_set = train_datagen.flow_from_directory(train_dir,target_size = (64, 64), batch_size = batch_size) #从文件路径中读取数据，路径中文件是按照分类进行存放的，读进来就会包含label信息
#val_set = val_datagen.flow_from_directory(validate, target_size = (64, 64), batch_size = batch_size)
#test_set = test_datagen.flow_from_directory(test_dir,
                                            #target_size = (64, 64),
                                            #batch_size = batch_size)
training_set=train_datagen.flow_from_dataframe(dataframe=x_train, directory="/content/shoe/data/Footwear", x_col="Image", y_col="Gender", class_mode="binary", target_size=(64,64), batch_size=50)
val_set=val_datagen.flow_from_dataframe(dataframe=x_val, directory="/content/shoe/data/Footwear", x_col="Image", y_col="Gender", class_mode="binary", target_size=(64,64), batch_size=50)
test_set=test_datagen.flow_from_dataframe(dataframe=X_test, directory="/content/shoe/data/Footwear", x_col="Image", y_col="Gender", class_mode="binary", target_size=(64,64), batch_size=50)
```

```python
from tensorflow.keras.datasets.cifar10 import load_data
from tensorflow.keras.applications.resnet import preprocess_input, ResNet50
import tensorflow.keras as K 
import tensorflow as tf
from keras import losses
#def preprocess_res(X, Y):
#    X_out = preprocess_input(X)
#    Y_out = K.utils.to_categorical(Y, 10)
#    return X_out, Y_out

#X_train, y_train = preprocess_res(x_train, y_train)
#X_test, y_test = preprocess_res(x_test, y_test)

res_model = ResNet50(include_top=False,
                      weights="imagenet",
                      input_tensor=K.Input(shape=(64, 64, 3)))

for layer in res_model.layers[:143]:
    layer.trainable = False

model = K.models.Sequential()
model.add(res_model)
model.add(K.layers.Flatten())
model.add(K.layers.Dropout(0.5))
model.add(K.layers.BatchNormalization())
model.add(K.layers.Dense(32, activation='relu'))
model.add(K.layers.BatchNormalization())
model.add(K.layers.Dense(32, activation='relu'))
model.add(K.layers.Dropout(0.5))
model.add(K.layers.Dense(1, activation='sigmoid'))

model.compile(loss=losses.BinaryCrossentropy(),
                  optimizer=K.optimizers.RMSprop(),
                  metrics=['accuracy'])

history = model.fit(training_set, validation_data = val_set, epochs = 60, verbose=1)
```

```python
test_loss, test_accuracy = model.evaluate(test_set)
```
